```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
ds = pd.read_csv('Speed Dating Data.csv', encoding = "ISO-8859-1",thousands=',')
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
ds2 = ds[['field','match']]
ds2
```

```python
ds2.groupby('field').count().sort_values(by=['match'],ascending=False)
```

```python
ds_nbmatch = ds2.groupby('field').sum().sort_values(by=['match'],ascending=False)
ds_total_cat = ds2.groupby('field').count().sort_values(by=['match'],ascending=False)
ds_pourcent = ds_nbmatch/ds_total_cat
```

```python
ds_top10 =ds_pourcent.sort_values(by='match',ascending=False).iloc[0:10,:]
```

```python
plt.figure(figsize=(20,10))

plt.bar(ds_top10.index,ds_top10['match'])
plt.show()
```

```python
ds['New_Income']= ds.income.apply(lambda x: "<20k" if x <20000
                                 else "20k<40k" if x < 40000
                                 else "40k<70k" if x <70000
                                else ">70k" if x <120000
                                 else "None")
```

```python
ds.New_Income
```

```python
ds_income_match = ds[['New_Income','match']].groupby('New_Income').sum().sort_values(by=['match'],ascending=False)
ds_incom_tot = ds[['New_Income','match']].groupby('New_Income').count().sort_values(by=['match'],ascending=False)
ds_income_pourcent = ds_income_match/ds_incom_tot
```

```python
print(ds_income_match,ds_incom_tot,ds_income_pourcent)
```
